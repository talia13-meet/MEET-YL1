```python
import csv 
f=open('cigarettes.csv', 'rb')
reader=csv.reader(f)
Tar_content=[]
Nicotine_content=[]
for row in reader:
    
    a=row[0]
    b=a.replace("\t"," ")
    c=b.split()
    d=c[1:]
    h=map(float, d)
    Tar_content.append(h[0])
    Nicotine_content.append(h[1])
print Tar_content
print Nicotine_content
  
    
  
        
   
f.close()

        
    
```

```python
from pylab import*
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title("cigarettes Tar and Nicotine content")
ax.set_xlabel("Tar content")
ax.set_ylabel("Nicotine content")              
scatter(Tar_content,Nicotine_content)
savefig("cig_scatter.png")
```
